# Performing EDA on the Dataset

## Question_1
### Which are the movies with the third lowest and third highest budget?

```python
largest_budget = df.nlargest(3, ['budget'])
largest_budget.tail(1)
```

```python
df4 = df[(df.budget != 0)]
df5 = df4.sort_values(by='budget', ascending=False, na_position='first')
df5.nsmallest(3, 'budget')



#df4.tail(3)
#df5.head(1)
```
